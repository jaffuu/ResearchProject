```python
from pycaret.classification import *
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
#performing feature engineering
# Fetch the dataset
abalone = fetch_ucirepo(id=1)

# Data (X = features, y = target)
X = abalone.data.features
y = abalone.data.targets

# Combine features and target into a single DataFrame for PyCaret
data = pd.concat([X, y], axis=1)

# Ensure the "Sex" column exists and set it as the target
data.rename(columns={"Sex": "target"}, inplace=True)

# Check initial data (optional)
print(data.head())

# --- Feature Engineering Section ---

# Encode the 'target' (Sex) column as a categorical variable
# This step ensures compatibility with PyCaret
data['target'] = data['target'].astype('category')

# Create a new feature: "Volume" (approximating abalone's physical volume)
# Volume = π * (Diameter/2)^2 * Height
data['Volume'] = np.pi * (data['Diameter'] / 2)**2 * data['Height']

# Create interaction features
data['Diameter_Height'] = data['Diameter'] * data['Height']
data['Weight_Ratio'] = data['Whole_weight'] / (data['Shell_weight'] + 1e-6)  # Avoid division by zero

# Normalize continuous features (e.g., weights, sizes)
scaler = MinMaxScaler()
scaled_columns = ['Length', 'Diameter', 'Height', 'Whole_weight', 'Shucked_weight', 'Viscera_weight', 'Shell_weight']
data[scaled_columns] = scaler.fit_transform(data[scaled_columns])

# Check data after feature engineering
print(data.head())

# --- PyCaret Setup ---
clf_setup = setup(data=data, target="target", session_id=123, preprocess=True)

# Compare all models
best_model = compare_models()

# Finalize the best model (train on entire dataset)
final_model = finalize_model(best_model)

# Evaluate the finalized model
evaluate_model(final_model)

# Save the model
save_model(final_model, "abalone_sex_multiclass_model")
```
